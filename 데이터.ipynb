```python
import pandas as pd
import os

# CSV 파일이 있는 디렉토리 경로 설정
csv_directory = "C:\Users\user\MINI4"

# 디렉토리 내 모든 CSV 파일 가져오기
csv_files = [f for f in os.listdir(csv_directory) if f.endswith('.csv')]

# CSV 파일을 담을 빈 데이터프레임 리스트 생성
dfs = []

# 각 CSV 파일을 순회하며 데이터프레임으로 불러오기
for csv_file in csv_files:
    file_path = os.path.join(csv_directory, csv_file)
    df = pd.read_csv(file_path)
    dfs.append(df)

# 모든 CSV 파일의 데이터프레임을 합치기 (옵션)
combined_df = pd.concat(dfs, ignore_index=True)

# 각 데이터프레임 출력
for i, df in enumerate(dfs):
    print(f"데이터프레임 {i + 1}:")
    print(df)

# 모든 CSV 파일의 데이터프레임을 출력 (옵션)
print("모든 데이터프레임 합침:")
print(combined_df)
```
